```python
#The columns required for my purpose are:-
#Average Cost for two
#Has Table booking
#Has Online delivery
#Price range
x = rest_data[['Average Cost for two','Has Table booking','Has Online delivery','Price range']]
```

```python
y = rest_data["Aggregate rating"]
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelEncoder = LabelEncoder()
```

```python
x["Has Table booking"] = labelEncoder.fit_transform(x["Has Table booking"])
```

```python
x["Has Online delivery"] = labelEncoder.fit_transform(x["Has Online delivery"])
```

```python
#Has table booking 0 is no an 1 is yes
#Has Online booking 0 is no and 1 is yess
```

```python
x.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(x)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
def modelresults(predictions):
    print("Mean absolute error on model is {}".format(mean_absolute_error(y_test, predictions)))
    print("Mean squared error on model is {}".format(np.sqrt(mean_squared_error(y_test, predictions))))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predictionFromLr = lr.predict(X_test)
```

```python
modelresults(predictionFromLr)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svmmodel = SVR()
```

```python
param_gridsvr = {"C" : [0.1, 0.5], "kernel": ["linear","poly"], "degree" : [ 2 ] }
```

```python
gridsvr = GridSearchCV(svmmodel, param_gridsvr)
```

```python
gridsvr.fit(X_train, y_train)
```

```python
gridsvr.predict(X_test)
```

```python
predsgridsvr = gridsvr.predict(X_test)
```

```python
modelresults(predsgridsvr)
```

```python
#both models performed similar
```

```python
param_grid = {
    "max_depth" : [10,30],
    "min_samples_leaf" : [1,2],
    "min_samples_split" : [2,5]
}
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
treemodel = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = treemodel, param_grid = param_grid)
```

```python
grid_tree.fit(X_train, y_train)
```

```python
treepredictions = grid_tree.predict(X_test)
```

```python
modelresults(treepredictions)
```
